```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data=pd.read_csv(r'https://raw.githubusercontent.com/amankharwal/Website-data/fb41da7f0f796d3a319e96a8751057b89ee6d225/weatherHistory.csv')
```

```python
data.head()
```

```python
data.columns
```

```python
col_req=['Formatted Date','Apparent Temperature (C)','Humidity']
data=data[col_req]
```

```python
data.head()
```

```python
data.shape
```

I have to convert formatted date into datetime format.

```python
data['Formatted Date']=pd.to_datetime(data['Formatted Date'],utc=True)
data=data.set_index('Formatted Date')
fdata=data[['Apparent Temperature (C)','Humidity']].resample('MS').mean()
fdata.head()
```

We will now plot the line graph to display Average Humidity and Average Apparent Temperature over 10 years(2006â€“2016). 

```python
plt.figure(figsize=(15,5))
plt.plot(fdata['Humidity'],label= 'Humidity',color='red',linestyle='dashed')
plt.plot(fdata['Apparent Temperature (C)'],label='Apparent Temperature (C)',color='blue')
plt.title('Variation Apparent temperature Vs c',fontsize=28)
plt.legend(loc=0,fontsize=12)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15);
```

We will now plot the line graph to display average temperature and humidity for  all 12 months over the 10 year period . Lets start from First month of the year.

1) january

```python
jan=fdata[fdata.index.month==1]
plt.figure(figsize=(15,5))
plt.plot(jan.loc['2006-04-01':'2016-04-01','Apparent Temperature (C)'],marker='o',linestyle='-',label='Apparent Temperature (C)',color='blue')
plt.plot(jan.loc['2006-04-01':'2016-04-01','Humidity'],marker='o',linestyle='-',label='Humidity',color='red')
plt.legend(loc='lower right',fontsize=15)
plt.xlabel('Month of January',fontsize=15)
plt.title('Humidity Vs Apparent temperature',fontsize=15);
```
